```python
import numpy as np
import itertools as it
import sys
#Total particles = A
# No. of fragements = F
# No. of particles in a cluster = Z_1,Z_2
A = int(input("Enter the total number of particles: "))
F = 2
Z_1 , Z_2 = input("Number of particles in each cluster: ").split()
#print("Z_1,Z_2:",Z_1,Z_2)
#identical pairs
Z_1 = int(Z_1)
Z_2 = int(Z_2)


z1 = Z_1
z2 = Z_2


#list of particles in cluster 1
Z_1ls = []
while Z_1 != 0:
    Z_1ls.append(Z_1)
    Z_1 = Z_1 -1
#print("Z_1ls:",Z_1ls)
Z_1ls.reverse()
#print("Z_1ls:",Z_1ls)
#list of particles in cluster 2
Z_2ls = []
while Z_2 != 0:
    Z_2ls.append(Z_2)
    Z_2 = Z_2 -1
#print("Z_2ls:",Z_2ls)

#Rename particles in cluster 2
Z_3ls = []
for i in range(len(Z_2ls)):
    a = Z_2ls[i] + z1
    Z_3ls.append(a)
Z_3ls.reverse()
#print("Z_2ls:",Z_3ls)

#identical pairs
#1_For IntBosePairs
idPairs =[]
id =  Z_1ls[0],Z_3ls[0]
idPairs.append(id)
#print("idPairs:",idPairs)


#2_For IntBoseTriplets
idTriples = []
idTriples.append(Z_1ls[0])
idTriples.append(Z_3ls[0])
#print("idTriples:",idTriples)



#Interacting Pairs
inpairs = []
intPairs = []
for i in range(len(Z_1ls)):
    for j in range(len(Z_3ls)):
        inpairs = Z_1ls[i], Z_3ls[j]
        intPairs.append(inpairs)
#print("intPairs",intPairs)

#Interacting Bose pairs
intPairs.remove(idPairs[0])
#print("IntBosPairs: ",intPairs)


#Interacting Triples
pairs =[]
pairs2 = []
for pair in it.combinations(Z_1ls,2):
    pairs.append(list(pair))
#print("pairs:",pairs)

#Interacting Triples
IntTriples = list(it.product(pairs, Z_3ls))
#print("IntTriples: ",IntTriples)

#Interacting Bose Triples
IntBosTriples =[]
for i in range(len(IntTriples)):
    if ((1 in IntTriples[i][0]) and (a ==  IntTriples[i][1])) :
        pass
    else:
        IntBosTriples.append(IntTriples[i])
#print("IntBosTriples: ",IntBosTriples)

#Interaction
#print("pairs:", intPairs)
#print("Triples", IntBosTriples)
```

```python
#####################################################################
#######################PART@2########################################

#construction of position vector
r = []
for i in range(1,A+1):
    i = str(i)
    g = "r"+i
    r.append(g)
#print("r vectors:",r)

#identity matrix
x = np.identity(A)
y = np.array(x)
#print(x)
#print("y:",y)




#COM of whole System
COM_whole = []
for i in range(A):
    COM_whole.append(1/A)
#print("COM_whole:",COM_whole)

#COM of first cluster
COM_Z1 =[]
for i in range(1,len(Z_1ls+Z_3ls)+1):
    if i in Z_1ls:
        COM_Z1.append(1/z1)
    else:
        COM_Z1.append(0)
#print("Rcom_Z1:", COM_Z1)

#COM of second cluster
COM_Z2 =[]
for i in range(1,len(Z_1ls+Z_3ls)+1):
    if i in Z_3ls:
        COM_Z2.append(1/z2)
    else:
        COM_Z2.append(0)
#print("Rcom_Z2:", COM_Z2)

#Position of (Z_1ls-1) particles  with respect to COM of first cluster
pcom = []
for i in range(len(Z_1ls)-1):
    if len(Z_1ls)-1 != 0:
        pcom.append(y[i] - COM_Z1)
    #print(pcom)
pcom = list(pcom)
#print("r_i = ",pcom)


#Position of (Z_3ls-1) particles  with respect to COM of first cluster
pcom2 = []
for i in range(len(Z_3ls)-1):
    if len(Z_3ls)-1 != 0:
        pcom2.append(y[Z_3ls[i]-1] - COM_Z2)
    #print(pcom2)
pcom2 = list(pcom2)
#print("r_j = ",pcom2)

#Distance between COM of both clusters

A_COM_Z1 = np.array(COM_Z1)
A_COM_Z2 = np.array(COM_Z2)
R_cm = np.subtract(A_COM_Z1,A_COM_Z2)
R_cm = list(R_cm)
#print("R_cm = ",R_cm)

#Single to cluster matrix
Single_to_Cluster = []
if len(pcom) != 0:
    for i in range(len(pcom)):
        Single_to_Cluster.append(list(pcom[i]))
if len(pcom2) != 0:
    for j in range(len(pcom2)):
        Single_to_Cluster.append(list(pcom2[j]))
Single_to_Cluster.append(R_cm)
Single_to_Cluster.append(COM_whole)

Single_to_Cluster = list(Single_to_Cluster)

m = np.reshape(Single_to_Cluster,(A,A))

#print("Single to cluster matrix: ",Single_to_Cluster)
SingleToCluster = Single_to_Cluster


#inverse of single to cluster
#np.linalg.inv(Single_to_Cluster)
#print("Inverse matrix: ",Single_to_Cluster)

```
